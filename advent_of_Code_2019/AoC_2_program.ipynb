### Part 2: 
Coding the computer to return a specific result

```python
def intcode(inp1, inp2):
    with open(os.path.join(sys.path[0], "AoC_2_program.txt"), "r") as f:
        df = pd.read_csv(f, sep=',', header = None)
    
    df.iloc[0][1] = inp1
    df.iloc[0][2] = inp2
    
    check = 0 
    position = 0
    while check == 0:
        opcode = df.iloc[0][position]
        num1 = df.iloc[0][position+1]
        num2 = df.iloc[0][position+2]
        store = df.iloc[0][position+3]

        if opcode == 1:
            val = df.iloc[0][num1]+df.iloc[0][num2]
        elif opcode == 2:
            val = df.iloc[0][num1]*df.iloc[0][num2]
        elif opcode == 99:
            check = 1
            break
        df.iloc[0][store] = val

        position +=4
    
    return df.loc[0][0]
```

determine what pair of inputs produces the output 19690720.

```python
##Inputs
noun = int(input('from 0 to 99'))
verb = int(input('from 0 to 99'))
print(intcode(noun, verb))
print('What is 100 * noun + verb? ', (100*noun)+verb)
```

```python
for noun in range(99+1):
    for verb in range(99+1):
        result = intcode(noun, verb)
        if result == 19690720:
            break
        print(noun, verb, result)
    else:
        # Continue if the inner loop wasn't broken.
        continue
    # Inner loop was broken, break the outer.
    break
    
print(noun, verb, (100*noun)+verb)
```
