Opcodes -  1 = add <br>
        -  2 = mulitply<br>
        - 99 = terminate<br>

before running the program, replace position 1 with the value 12 and replace position 2 with the value 2. What value is left at position 0 after the program halts?

```python
df.iloc[0][1] = 12
df.iloc[0][2] = 2
```

```python
df
```

```python
check = 0 
position = 0
while check == 0:
    opcode = df.iloc[0][position]
    num1 = df.iloc[0][position+1]
    num2 = df.iloc[0][position+2]
    store = df.iloc[0][position+3]
    
    if opcode == 1:
        val = df.iloc[0][num1]+df.iloc[0][num2]
    elif opcode == 2:
        val = df.iloc[0][num1]*df.iloc[0][num2]
    elif opcode == 99:
        check = 1
        print('encountered halt', opcode, num1, num2, store, val)
        break
    
    print(opcode, num1, num2, store, val)
    
    df.iloc[0][store] = val
    print(df.values)
    position +=4
```
