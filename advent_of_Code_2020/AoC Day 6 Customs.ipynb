```python
import pandas as pd
import os, sys
```

```python
with open(os.path.join(sys.path[0], "AoC_6_CustomCustoms.txt"), "r") as f:
    contents = f.read()
    custgroups = contents.split("\n\n")
```

```python
custgroups
```

```python
question_sum = 0 
for custgroup in custgroups:
    group_mod = custgroup.replace('\n','').strip()
    current_sum = len(set(group_mod))
    question_sum = question_sum + current_sum

print('Total questions answered "yes" and summed:', question_sum)
```

```python
question_sum = 0 

for custgroup in custgroups:
    group_mod = custgroup.replace('\n','').strip()
    index = set(group_mod)
    group = custgroup.split('\n')
    group = list(filter(None, group))
    group_size = len(group)
    
    ##print(group_size, len(index), group_mod, index, group)
    
    occurance = []
    
    for char in index:
        counter = group_mod.count(char)
        if counter == group_size:
            occurance.append(char)
            
    ##print(group_size, occurance, len(occurance))
    question_sum += len(occurance)


print('Total questions answered "yes" by all members in a group and summed:', question_sum)
```
