```python
import pandas as pd
import os, sys
```

```python
with open(os.path.join(sys.path[0], "AoC_3_Trees.txt"), "r") as f: ##open file
    lines = f.read().splitlines()
```

```python
trees = []
for line in lines: ##create list of list for each line
    trees.append(list(line))
```

```python
trees
```

```python
df = pd.DataFrame(trees) ##into dataframe
```

```python
df
```

```python
##Part 1 find the number of tree strikes with the three right, one down step pattern
##start location (0,0)
##motion = +3, +1 three right, one down
##left right 0 to 30 and it repeats to the right. So column 30 wraps back to zero
##top bottom 0 to 322
check = 0
start_row = 0
start_col = 0
step_right = 3
step_down = 1
tree_sum = 0
clear_sum = 0
x_pos = start_col
y_pos = start_row

while check <=(322-step_down):
    x_pos += step_right
    if x_pos > 30: x_pos -= len(df. columns)
    
    y_pos += step_down
    
    if df.loc[y_pos,x_pos] == '#':
        tree_sum += 1
    elif df.loc[y_pos,x_pos] == '.':
        clear_sum += 1
    
    check = y_pos

print('Trees hit: ',tree_sum, ' Open space hit: ',clear_sum)
```

```python
##Part 2 try multiple step patterns and multiply the tree hits together
##start location (0,0)
##motion is variable with right and down combos [[1,1],[3,1],[5,1],[7,1],[1,2]]
##left right 0 to 30 and it repeats to the right. So column 30 wraps back to zero
##top bottom 0 to 322

start_row = 0
start_col = 0
step_combo = [[1,1],[3,1],[5,1],[7,1],[1,2]]
tree_sum = 0
clear_sum = 0
multi = 1

for combo in step_combo:
    step_right = combo[0]
    step_down = combo[1]
    check = 0
    tree_sum = 0
    clear_sum = 0
    x_pos = start_col
    y_pos = start_row
    
    while check <=(322-step_down):
        x_pos += step_right
        if x_pos > 30: x_pos -= len(df. columns)

        y_pos += step_down

        if df.loc[y_pos,x_pos] == '#':
            tree_sum += 1
        elif df.loc[y_pos,x_pos] == '.':
            clear_sum += 1

        check = y_pos

    print('Step style: ',combo, ' Trees hit: ',tree_sum, ' Open space hit: ',clear_sum)
    multi = multi * tree_sum
print('Trees hit for all cases multiplied together: ',multi)
```
