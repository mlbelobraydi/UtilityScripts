```python
import pandas as pd
import os, sys
```

```python
with open(os.path.join(sys.path[0], "AoC_7_Haversacks.txt"), "r") as f:
    bag_rules = f.read().split('\n')
```

How many bag colors can eventually contain at least one shiny gold bag?<br>
Due to recent aviation regulations, many rules (your puzzle input) are being enforced about bags and their contents; bags must be color-coded and must contain specific quantities of other color-coded bags. Apparently, nobody responsible for these regulations considered how long they would take to enforce!

```python
df = pd.DataFrame()
col = ('main_bag', 'contained_bags')

for rule in bag_rules:
    line = rule.split(' bags contain ')
    main_bag = line[0]
    contain_bag = line[1].split(',')
    contain_bag = [sub.replace('bags', '') for sub in contain_bag] 
    contain_bag = [sub.replace('bag', '') for sub in contain_bag] 
    contain_bag = [sub.replace('.', '') for sub in contain_bag] 
    allbags = []
    for contained in contain_bag:
        contained = contained.strip()
        entry = contained.split(" ", 1)

        if entry[0] == 'no':
            number = 0
            bag_type = 'NoBag'
        else:
            number = int(entry[0])
            bag_type = str(entry[1])
        allbags.append((bag_type,number))
    
    df = df.append([(main_bag,allbags)], ignore_index = True)

df.columns = col
```

```python
def BagRuleList(bag,bagdf):
    result = 0
    cont_bags = df.loc[df['main_bag'] == bag, 'contained_bags'].tolist()
    if cont_bags:
        if cont_bags[0] != 'NoBag':
#             print(cont_bags[0])
            for bg in cont_bags[0]:
                if bg[0] == 'shiny gold':
                    result += int(bg[1])
#                     print(bag, bg)
                found = bagdf.main_bag[bagdf['main_bag'].str.contains(bg[0])].count()
#                 print(found)
                if found > 0:
                    result = result + BagRuleList(bg[0],bagdf)
    shinyCt = result
    return shinyCt
```

```python
bagList = df['main_bag'].values
```

```python
ct = 0
bagCt = 0
for bag in bagList:
    bagCt +=1
    shinyCt = BagRuleList(bag,df)
    if shinyCt >0: 
        ct+=1
        print(bag, shinyCt, ct)
print('Total rules holding "shiny gold" bags: ', ct, ' out of ', bagCt, ' rules.')
```

Part 2: How many individual bags are required inside your single shiny gold bag?

```python
df.loc[df['main_bag'] == 'shiny gold', 'contained_bags'].tolist()
```

```python
def ShinyGoldContained(bag, bagMulti, bagdf):
    result = 0
    cont_bags = df.loc[df['main_bag'] == bag, 'contained_bags'].tolist()
    if cont_bags:
        if cont_bags[0] != 'NoBag':
            for bg in cont_bags[0]:
                result += bg[1]
                bgMulti = bg[1]
                found = bagdf.main_bag[bagdf['main_bag'].str.contains(bg[0])].count()
                if found > 0:
                    result = result + ShinyGoldContained(bg[0], bg[1], bagdf)
    totalbags = bagMulti*result
    return totalbags
```

```python
bagSum = 0
for baglst in df.loc[df['main_bag'] == 'shiny gold', 'contained_bags'].tolist():
    for bag in baglst:
        bag_contain = bag[1] + ShinyGoldContained(bag[0], bag[1], df)
        bagSum += bag_contain 
        print(bag, bag_contain)
print('Total bags held in single "shiny gold" bag: ', bagSum)
```
