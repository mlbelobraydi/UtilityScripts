```python
import pandas as pd
```

```python
import os, sys
```

```python
with open(os.path.join(sys.path[0], "AoC_2_pw.txt"), "r") as f:
    df= pd.read_csv(f, sep=' ', header = None)
```

```python
df
```

```python
df[['low','high']] = df[0].str.split('-',expand=True).astype('int32')
```

```python
df['char'] = df[1].str.replace(r':', '')
```

```python
df
```

```python
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'ct'] = row[2].count(row.char)
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'check'] =  (row['ct'] >= row['low']) & (row['ct'] <= row['high'])
        
for i, row in df.iterrows():
    if row.char in row[2]:        
        df.loc[i,'first_pos'] = row[2][row['low']-1] == row['char']
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'second_neg'] = row[2][row['high']-1] != row['char']
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'check2'] = row['first_pos'] == row['second_neg'] == True
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'first_neg'] = row[2][row['low']-1] != row['char']
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'second_pos'] = row[2][row['high']-1] == row['char']
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'check3'] = row['first_neg'] == row['second_pos'] == True
        
for i, row in df.iterrows():
    if row.char in row[2]:
        df.loc[i,'check4'] = (row['check2'] == True) or (row['check3'] == True)
```
